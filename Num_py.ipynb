```python
#properties
arr=np.array([[1,2,3,4,5],[2,5,8,7,9]])
print(arr.shape)#dimensions =mxn
print(arr.size)
print(arr.dtype)
print(arr.ndim)
```

```python
#operations
arr = np.array([[1,2,3],[4,5,6]])
print(arr,arr.shape)

#reshaping

reshaped = arr.reshape((3,2))
print(reshaped,reshaped.shape)

#flatened

flattened = arr.flatten()
print(flattened,flattened.shape)
```

```python
#indexing
arr=np.array([1,2,3,4,5])
print(arr[0])
print(arr[2])
```

```python
arr = np.array([[1,2,3],[4,5,6]])
print(arr[1][1])
```

```python
#fancy indexing
arr=np.array([1,2,3,4,5,6,7])
idx=[0,1,3,5,6]
print(arr[idx])
```

```python
#slicing
arr = np.array([1,2,3,4,5])
#[start,end,step]
print(arr[1:4])
print(arr[1:])
print(arr[:4])
```
